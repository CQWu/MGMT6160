```python
# We handle the exceptions with the original error types.
# It still looks rather ugly.
def factorial(n):
    try:    
        if n == 0:
            return 1
        else:
            return n*factorial(n-1)
    except (RuntimeError, TypeError) as e:
        print e
```

```python
factorial(-2) #RuntimeError
```

```python
factorial('x')    #TypeError
```

```python
factorial(2.4) #RuntimeError
```
